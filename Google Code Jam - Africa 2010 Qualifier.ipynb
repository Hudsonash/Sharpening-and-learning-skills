##### Problem A: Store Credit
Recieve c credit at a local store and would like to purchase two objects that use the entire store credit.  Return a case of 

```python
## store_credit(credit, total_items, item_prices) takes a credit
##  value, the total items in the store, and a list of the 
##  prices in the store.  It returns the index of two items 
##  in the list which add up to the credit.
## Float, Nat, Float -> str

def store_credit(credit, total_items, item_prices):
    loop_count1 = 0 
    for item in item_prices:
        loop_count2 = 0 
        for item_add in item_prices:
            if loop_count2 == loop_count1:
                pass
            elif (item + item_add) == credit:
                return "Case #: {0} {1}".format(loop_count1 + 1,
                                                loop_count2 + 1)
            else: pass
            loop_count2 = loop_count2 + 1
        loop_count1 = loop_count1 + 1
    
    return "No Possible Combination"


    
            
            
            
```

##### Problem 2: Reverse Words
Given a list of space separated words, reverse the order of the words.  Each line of text contains L letters and W words. Exactly one space between each pair of consecutive words 

```python
## reverse_words(string) takes a sentence and 
##  reverses the order of the words
## Str -> Str
## Example: reverse_words('a b c d') -> 'd c b a'

def reverse_words(string):
    string = list(reversed(string.split(' ')))
    loop_count = 0 
    new_sen = ''
    for word in string:
        new_sen = new_sen + word + ' '
        
    if len(new_sen) < 2:
        return 'Case #: {0}'.format(new_sen)
    else: return 'Case #: {0}'.format(new_sen[:-1])
    
```

```python
## preparing the testing for the reverse_words_test
small_test = open('B-small-practice.in', 'r')
small_words = small_test.read().split('\n')[:-1]
small_test.close()

large_test = open('B-large-practice.in', 'r')
large_words = large_test.read().split('\n')[:-1]
large_test.close()


## reverse_words_test(lst_of_sen) reads an list of tests
##  and prints the results given by reverse_words
def reverse_words_test(lst_of_sen):
    tests = int(lst_of_sen[0])
    test_count = 1
    while tests >= test_count:
        sen_to_prnt = reverse_words(lst_of_sen[test_count])
        print(sen_to_prnt[:6] + str(test_count) + sen_to_prnt[6:])
        test_count = test_count + 1

# small and large tests
reverse_words_test(small_words)
reverse_words_test(large_words)
```

##### Problem 3: T9 Spelling
Input a sentence and convert to how it would be represented on a flip-phone.  Ex. C = 111, or U = 88.  The space character indicates a pause in button pressing.  In the test, only lowercase letters will be used. 


```python
## A dictionary of conversions from letter to number 
char_to_num_dct = {'a': '2', 'b': '22', 'c': '222',
                   'd': '3', 'e': '33', 'f': '333',
                   'g': '4', 'h': '44', 'i': '444',
                   'j': '5', 'k': '55', 'l': '555',
                   'm': '6', 'n': '66', 'o': '666',
                   'p': '7', 'q': '77', 'r': '777', 's': '7777',
                   't': '8', 'u': '88', 'v': '888', 
                   'w': '9', 'x': '99', 'y': '999', 'z': '9999',
                   ' ': '0'}
## t9_spelling(string) takes a string and converts the 
##  characters from the string into phone digits
## Str -> Str
## Example t9_spelling('yes') -> '999337777'
def t9_spelling(string):
    number_str = ' '
    for letter in string:
        if char_to_num_dct[letter][0] == number_str[-1]:
            number_str = number_str + ' ' + char_to_num_dct[letter]
        else: number_str = number_str + char_to_num_dct[letter]
    return number_str[1:]
```

```python
## preparing test files for t9_test
small_test_c = open('C-small-practice.in', 'r')
small_words_c = small_test_c.read().split('\n')[:-1]
small_test_c.close()

large_test_c = open('C-large-practice.in', 'r')
large_words_c = large_test_c.read().split('\n')[:-1]
large_test_c.close()

## t9_test(lst_of_sen) takes a list of sentences and 
##  prints the test result for each test. 
def t9_test(lst_of_sen):
    tests = int(lst_of_sen[0])
    test_count = 1
    while tests >= test_count:
        sen_to_prnt = t9_spelling(lst_of_sen[test_count])
        print('Case #{0}: '.format(test_count) + sen_to_prnt)
        test_count = test_count + 1

t9_test(small_words_c)
t9_test(large_words_c)
```
