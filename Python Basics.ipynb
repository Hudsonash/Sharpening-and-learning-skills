```python
def reverse_string(string):
    length = len(string) 
    reverse_string = ''
    counter = -1
    while (counter * -1) <= length:
        reverse_string = reverse_string + string[counter]
        counter = counter - 1 
    return reverse_string
    
    
```

```python
def reverse_string2():
    user_word = input('Please enter word here: ')
    if type(user_word) != str:
        print('Sorry, that is not a string. It was type {0}'.format(type(user_word)))
        user_word = input('Please re-enter word here: ')
    print('Reversing... \n')
    print(reverse_string(user_word) + '\n')
    print('The word has been reversed. Thank you')
```
