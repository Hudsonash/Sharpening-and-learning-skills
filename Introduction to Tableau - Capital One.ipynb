**_ What is Tableau?_**  
Tableau is primarily a data visualization tool to be used for analytics and business intelligence. It is a market leader in BI products, with five main products.  A desktop product for individual use, a server for organizational use, a cloud-based server for business intelligence on the cloud, and Tableau Public to publish data visualizations online.  

[Download Tableau Public Here](https://public.tableau.com/en-us/s/download/)

The Tableau Public is the free software whereas the desktop, organizational, and cloud-based servers cost money.  The essential difference is what forms of data are accessed and where the visualizations are stored.  The premium versions allow for many different forms of data to be visualized.

### Tableau Sheets 
A Tableau sheet seperates the data from the table into two sections: The first is _dimensions_, and the second is measures.  Dimensions and measures are then dragged into a tableau sheet, under either columns or rows to create a table.  From the sheet, graphs are automatically generated which display Tableau's suggested visualizations.  Below is an image of a Tableau Sheet, where dimensions and measures can be dragged and dropped. 

![Sheet view from Tableau](https://raw.githubusercontent.com/Hudsonash/Images-for-GitHub/master/Tableau-sheet.png)


Once the information is entered, several modifications can be made to the visualization.  The modifications are:
* **Colour**: The colour of the data points, bars, pie slices, etc. can be changed
* **Graph Type**: Changes the way the data is represented, whether it be bubbles, bars, or lines
* **Size**: Changes the size of the selected graph type
* **Label**: Changes the labels on the graph

Below is the image of where the visualization modifications can be completed along with the selction of dimensions and measures

![Sheet view from Tableau](https://raw.githubusercontent.com/Hudsonash/Images-for-GitHub/master/Tableau-cols-rows-effects.png)

### Uploading to Tableau Public
For Tableau Public, a visualization must be saved directly to the web to be accessed through the Tableau website.  
It is as simple as hitting save on the Tableau Public program, and entering a quick description of the graph on the website.  A url is then created for the publication to be viewed by the world!
